```
X = np.load('IF_data.npz')['arr_0']
Y = np.load('IF_data.npz')['arr_1']
print('the shape of X is : {}'.format(X.shape))
print('the shape of Y is : {}'.format(Y.shape))
print('egrg:{}'.format(Y[4]))
y = Y[4]
print('the shape of Y is : {}'.format(y.shape))
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state = 0)

print('the shape of Y train is : {}'.format(Y_train.shape))
print('the shape of X train is : {}'.format(X_train.shape))
opt_svr_param = {'C': 0.04, 'degree': 2, 'epsilon': 0.0015, 'gamma': 11}
svr1 = SVR(**opt_svr_param)

svr1 = SVR(kernel='poly', degree=2, gamma=11,epsilon=.0015, C=.04 )
svr1.fit(X_train,Y_train)

svr2=SVR(**opt_svr_param)
svr2=SVR(kernel='rbf', degree=3, gamma=11,epsilon=.15, C=.04 )
svr2.fit(X_train,Y_train)
```

```
ns_list =[1,94]
svc_sigmoid_scores = []
svc_poly_scores = []
svc_rbf_scores = []

svc_sigmoid_clf = SVC(kernel='sigmoid')
svc_poly_clf = SVC(kernel='poly')
svc_rbf_clf = SVC(kernel='rbf')

n_avg = 20                ## Number of times to average

for ns in ns_list:        ## For different subset-sizes we try training our estimator
  svc_e_in_sigmoid = 0
  svc_e_out_sigmoid = 0               ## For each, we record the in and out score.
  
  svc_e_in_poly = 0
  svc_e_out_poly = 0  
  
  svc_e_in_rbf = 0
  svc_e_out_rbf = 0  
  
  print(ns)

  for i in range(n_avg):
  
    X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state = 0)

    svc_sigmoid_clf.fit(X_train,Y_train)
    svc_poly_clf.fit(X_train,Y_train)
    svc_rbf_clf.fit(X_train,Y_train)

    svc_e_in_sigmoid  += svc_sigmoid_clf.score(X_train, Y_train )
    svc_e_out_sigmoid += svc_sigmoid_clf.score(X_test, Y_test )
    
    svc_e_in_poly  += svc_poly_clf.score(X_train, Y_train )
    svc_e_out_poly += svc_poly_clf.score(X_test, Y_test )
    
    svc_e_in_rbf  += svc_rbf_clf.score(X_train, Y_train )
    svc_e_out_rbf += svc_rbf_clf.score(X_test, Y_test )

  
  svc_sigmoid_scores += [[svc_e_in_sigmoid/n_avg, svc_e_out_sigmoid/n_avg]]
  svc_poly_scores += [[svc_e_in_poly/n_avg, svc_e_out_poly/n_avg]]
  svc_rbf_scores += [[svc_e_in_rbf/n_avg, svc_e_out_rbf/n_avg]]
  
svc_sigmoid_scores = np.array(svc_sigmoid_scores)
svc_poly_scores = np.array(svc_poly_scores) 
svc_rbf_scores = np.array(svc_rbf_scores) 


fig = plt.figure(figsize=(8,6))
xlabel = 'Sample Size'
ylabel = 'Accuracy'

svc_sigmoid_title = 'Learning curve for classification with SVC sigmoid'
svc_poly_title = 'Learning curve for classification with SVC poly'
svc_rbf_title = 'Learning curve for classification with SVC rbf'

ax1 = fig.add_subplot(111, xlabel=xlabel, ylabel=ylabel, title=svc_sigmoid_title)


ax1.plot(ns_list, svc_sigmoid_scores[:,0], label = 'Training Score')
ax1.plot(ns_list, svc_sigmoid_scores[:,1], label = 'Validation Score')
ax1.legend(loc=0)


plt.show()
```
